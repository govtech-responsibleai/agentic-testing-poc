```python
import json
import pandas as pd
```

```python
from pathlib import Path
results_set_dir = Path("/home/watsonchua/sls-lea-evaluation/data/eval/results_set")
results_set_files = list(results_set_dir.glob("*"))
len(results_set_files)
```

```python
results_map = {}

for file in results_set_files:
    print(file.stem)
    with open(file, "r") as f:
        data = json.load(f)
    
    for key, value in data.items():
        if key not in results_map:
            results_map[key] = {}
        # print(key)
        # print("\n\n")
        for v in value:
            if "result" in v:
                cc = (v["result"]["claims_check"])
                for claim in cc:
                    if "hallucination_check" in claim:
                        if "hallucination_check" not in results_map[key]:
                            results_map[key]["hallucination_check"] = []
                        results_map[key]["hallucination_check"].append(claim["hallucination_check"]["score"])
                        # print("Hallucination check: ", claim["hallucination_check"]["score"])
                    if "factuality_check" in claim:
                        if "factuality_check" not in results_map[key]:
                            results_map[key]["factuality_check"] = []
                        results_map[key]["factuality_check"].append(claim["factuality_check"]["score"])
                        # print("Factuality check: ", claim["factuality_check"]["score"])
                    
                    
                    


```

```python
from collections import Counter
records = []
for key, value in results_map.items():
    print(key)
    for k, v in value.items():
        print(k)
        print(Counter(v))
        records.append({"category":key, "check_type": k, **Counter(v)})
```

```python
df_results = pd.DataFrame(records)
```
